```python
# Function to download and extract datasets
def download_and_extract(url, zip_dir, extract_dir):
    if not os.path.exists(extract_dir):
        if not os.path.exists(zip_dir):
            !wget -O $zip_dir $url
        with zipfile.ZipFile(zip_dir, "r") as zip_ref:
            zip_ref.extractall()
        os.remove(zip_dir)
```

```python
# URLs for Visual Genome datasets
images1 = "https://cs.stanford.edu/people/rak248/VG_100K_2/images.zip"
images1_dir = "images"
images1_zip_dir = "images.zip"

images2 = "https://cs.stanford.edu/people/rak248/VG_100K_2/images2.zip"
images2_dir = "images2"
images2_zip_dir = "images2.zip"

# Download and extract datasets
download_and_extract(images1, images1_zip_dir, images1_dir)
download_and_extract(images2, images2_zip_dir, images2_dir)

# List the contents of the visual_genome directory
%ls
```
